# 1. Get top 20 players in VORP for each season from 2006 to 2022

```python
o6_o7 = pd.read_csv('data/2006-07.csv')
o7_o8 = pd.read_csv('data/2007-08.csv')
o8_o9 = pd.read_csv('data/2008-09.csv')
o9_o10 = pd.read_csv('data/2009-10.csv')
o10_o11 = pd.read_csv('data/2010-11.csv')
o11_o12 = pd.read_csv('data/2011-12.csv')
o12_o13 = pd.read_csv('data/2012-13.csv')
o13_o14 = pd.read_csv('data/2013-14.csv')
o14_o15 = pd.read_csv('data/2014-15.csv')
o15_16 = pd.read_csv('data/2015-16.csv')
o16_17 = pd.read_csv('data/2016-17.csv')
o17_18 = pd.read_csv('data/2017-18.csv')
o18_19 = pd.read_csv('data/2018-19.csv')
o19_20 = pd.read_csv('data/2019-20.csv')
o20_21 = pd.read_csv('data/2020-21.csv')
o21_22 = pd.read_csv('data/2021-22.csv')
```

```python
seasons = [o6_o7, o7_o8, o8_o9, o9_o10, o10_o11, o11_o12, o12_o13, o13_o14, 
           o14_o15, o15_16, o16_17, o17_18, o18_19, o19_20, o20_21, o21_22]
for i, season_df in enumerate(seasons):
    season_df['Season'] = f"{2006 + i}-{2007 + i}"

all_players = pd.concat(seasons, ignore_index=True)
all_players = all_players.drop(columns=['Awards', 'Player-additional', 'Rk'])
all_players.to_csv('data/all_players.csv', index=False)

all_players.head(10)
```

# 2. Getting the shared numeric columns across all dataframes

```python
numeric_cols_all = set(all_players.select_dtypes(include=[np.number]).columns)
numeric_cols_luke = set(luke_combined_data.select_dtypes(include=[np.number]).columns)
numeric_cols_ad = set(AD_combined_data.select_dtypes(include=[np.number]).columns)


common_numeric_cols = sorted(list(numeric_cols_all.intersection(numeric_cols_luke, numeric_cols_ad)))
print("Common numeric columns:", common_numeric_cols)


all_players_common = all_players.loc[:, ~all_players.columns.duplicated()]
luke_common = luke_combined_data.loc[:, ~luke_combined_data.columns.duplicated()]
ad_common = AD_combined_data.loc[:, ~AD_combined_data.columns.duplicated()]

all_players_numeric = all_players_common[common_numeric_cols]
luke_numeric = luke_common[common_numeric_cols]
ad_numeric = ad_common[common_numeric_cols]
```

# 3. Getting Luka and Ad's current seasons, combining the data, and setting a common scalar 

```python
luka_current = luke_numeric.iloc[-1:] 
ad_current = ad_numeric.iloc[-1:] 

combined_data = np.vstack([luka_current.values, ad_current.values, all_players_numeric.values]) ## https://www.geeksforgeeks.org/numpy-vstack-in-python/

scaler = StandardScaler()
combined_scaled = scaler.fit_transform(combined_data)
```

# 4. Running the k-means algorithm

```python
n_clusters = 10  # You can adjust this parameter
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(combined_scaled)
```

```python
luka_label = labels[0]
ad_label = labels[1]

print("Luka's label:", luka_label)
print("AD's label:", ad_label)
```

# 5. Clustering for Luka!

```python
all_players_scaled = combined_scaled[2:]
luka_center = kmeans.cluster_centers_[luka_label].reshape(1, -1) ## center of Luka's cluster
luka_distances = euclidean_distances(all_players_scaled, luka_center).flatten() ##distance from Luka to every other player
luka_cluster_mask = (labels[2:] == luka_label)
luka_cluster_indices = np.where(luka_cluster_mask)[0]
luka_cluster_distances = luka_distances[luka_cluster_mask]
top10_luka_indices_rel = luka_cluster_indices[np.argsort(luka_cluster_distances)[:25]]
luka_similar_seasons = all_players.iloc[top10_luka_indices_rel]

print("20 Most Similar Seasons to Luka's Current Season:")
print(luka_similar_seasons[['Player', 'Season']].to_string(index=False))
```

# 6. K-nearest for AD!

```python
ad_scaled = scaler.transform(ad_current) 
ad_distances = euclidean_distances(all_players_scaled, ad_scaled).flatten()
top10_ad_full_indices = np.argsort(ad_distances)[:25]
ad_similar_seasons_full = all_players.iloc[top10_ad_full_indices]
print("\nTop 20 Similar Seasons to AD's Current Season (Nearest Neighbors):")
print(ad_similar_seasons_full[['Player', 'Season']].to_string(index=False))
```
