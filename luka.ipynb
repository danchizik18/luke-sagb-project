# 3. Get top 20 players in VORP for each season from 2006 to 2022

```python
o6_o7 = pd.read_csv('data/2006-07.csv')
o7_o8 = pd.read_csv('data/2007-08.csv')
o8_o9 = pd.read_csv('data/2008-09.csv')
o9_o10 = pd.read_csv('data/2009-10.csv')
o10_o11 = pd.read_csv('data/2010-11.csv')
o11_o12 = pd.read_csv('data/2011-12.csv')
o12_o13 = pd.read_csv('data/2012-13.csv')
o13_o14 = pd.read_csv('data/2013-14.csv')
o14_o15 = pd.read_csv('data/2014-15.csv')
o15_16 = pd.read_csv('data/2015-16.csv')
o16_17 = pd.read_csv('data/2016-17.csv')
o17_18 = pd.read_csv('data/2017-18.csv')
o18_19 = pd.read_csv('data/2018-19.csv')
o19_20 = pd.read_csv('data/2019-20.csv')
o20_21 = pd.read_csv('data/2020-21.csv')
o21_22 = pd.read_csv('data/2021-22.csv')
```

```python
seasons = [o6_o7, o7_o8, o8_o9, o9_o10, o10_o11, o11_o12, o12_o13, o13_o14, 
           o14_o15, o15_16, o16_17, o17_18, o18_19, o19_20, o20_21, o21_22]
for i, season_df in enumerate(seasons):
    season_df['Season'] = f"{2006 + i}-{2007 + i}"

all_players = pd.concat(seasons, ignore_index=True)
all_players = all_players.drop(columns=['Awards', 'Player-additional', 'Rk'])
all_players.to_csv('data/all_players.csv', index=False)

all_players.head()
```
