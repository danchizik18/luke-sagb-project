```python
# Base URLs
index_url = "https://www.basketball-reference.com/players/"
base_player_url = "https://www.basketball-reference.com"

# Headers to mimic a real browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Function to get player stats
def get_player_stats(player_link):
    try:
        response = requests.get(player_link, headers=headers)
        if response.status_code != 200:
            print(f"Skipping {player_link} (status code: {response.status_code})")
            return None
        
        soup = BeautifulSoup(response.text, "html.parser")

        # Get player name safely
        name_tag = soup.find("h1", itemprop="name")
        name = name_tag.text.strip() if name_tag else "Unknown"

        # Find the career totals row
        career_row = soup.find("tfoot").find("tr") if soup.find("tfoot") else None
        if not career_row:
            print(f"Skipping {name} (no career totals found)")
            return None
        
        # Extract key stats
        def get_stat(stat_name):
            stat_cell = career_row.find("td", {"data-stat": stat_name})
            return stat_cell.text.strip() if stat_cell else "N/A"

        pts = get_stat("pts")
        reb = get_stat("trb")
        ast = get_stat("ast")
        vorp = get_stat("vorp")
        per = get_stat("per")

        return [name, player_link, pts, reb, ast, vorp, per]

    except Exception as e:
        print(f"Error processing {player_link}: {e}")
        return None

# Get player links from index pages
player_links = []
for letter in "abcdefghijklmnopqrstuvwxyz":
    response = requests.get(index_url + letter + "/", headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    
    table = soup.find("table", {"id": "players"})
    if table:
        for row in table.find("tbody").find_all("tr"):
            cols = row.find_all("td")
            if cols:
                link_tag = cols[0].find("a")
                link = base_player_url + link_tag["href"] if link_tag else None
                if link:
                    player_links.append(link)
    
    time.sleep(1)  # Be polite and avoid getting blocked

# Scrape stats for each player
all_stats = []
for link in player_links:
    player_stats = get_player_stats(link)
    if player_stats is not None:
        all_stats.append(player_stats)
    
    time.sleep(1)  # Avoid rate-limiting

# Create DataFrame
players_stats_df = pd.DataFrame(all_stats, columns=["Name", "Link", "PTS", "REB", "AST", "VORP", "PER"])

# Save to CSV
players_stats_df.to_csv("all_nba_players_stats.csv", index=False)

print(players_stats_df.head())  # Preview the data
```
