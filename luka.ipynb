# 3. Get top 20 players in VORP for each season from 2006 to 2022

```python
o8_o9 = pd.read_csv('data/2008-09.csv')
o6_o7 = pd.read_csv('data/2006-07.csv')
o10_o11 = pd.read_csv('data/2010-11.csv')
o9_o10 = pd.read_csv('data/2009-10.csv')
o11_o12 = pd.read_csv('data/2011-12.csv')
o12_o13 = pd.read_csv('data/2012-13.csv')
o13_o14 = pd.read_csv('data/2013-14.csv')
o14_o15 = pd.read_csv('data/2014-15.csv')
o15_16 = pd.read_csv('data/2015-16.csv')
```

```python
# List of DataFrames for each season (e.g., 'o0_o1' for 2000-2001, 'o1_o2' for 2001-2002, etc.)
seasons_data = [
    o0_o1, o1_o2, o2_o3, o3_o4, o4_o5, o5_o6, o6_o7, o7_o8, o8_o9, o9_o10,
    o10_o11, o11_o12, o12_o13, o13_o14, o14_o15, o15_o16, o16_o17, o17_o18, o18_o19, o19_o20,
    o20_o21, o21_o22
]

# List of season names corresponding to the DataFrames (you can adjust these if the names are different)
season_names = [
    "o0_o1", "o1_o2", "o2_o3", "o3_o4", "o4_o5", "o5_o6", "o6_o7", "o7_o8", "o8_o9", "o9_o10",
    "o10_o11", "o11_o12", "o12_o13", "o13_o14", "o14_o15", "o15_o16", "o16_o17", "o17_o18", "o18_o19", "o19_o20",
    "o20_o21", "o21_o22"
]

# Add the 'Season' column to each DataFrame
for season_df, season_name in zip(seasons_data, season_names):
    season_df['Season'] = season_name

# Concatenate all DataFrames into one large DataFrame
all_seasons_df = pd.concat(seasons_data, ignore_index=True)

# Display the first few rows of the merged DataFrame
print(all_seasons_df.head())
```
